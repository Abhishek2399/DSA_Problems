```python
from typing import *

def twoSum(nums: List[int], target: int) -> List[int]:
    idx_dict = dict()
    for i in range(len(nums)):
        if (target - nums[i]) in idx_dict:
            return [idx_dict[target - nums[i]], i]
        else:
            idx_dict[nums[i]] = i
```

```python
def isAnagram(s: str, t: str) -> bool:
    if len(s) != len(t) : return False
    counter = dict()

    for char in s:
        counter[char] = counter.get(char, 0) + 1

    for char in t:
        if char not in counter or counter.get(char, 0) == 0:
            return False
        counter[char] -= 1

    return True
```

```python
def groupAnagrams(strs: List[str]) -> List[List[str]]:
    groups = []
    for word in strs:
        anagrams = []
        for word2 in strs:
            if isAnagram(s=word, t=word2) and (word != word2):
                print(f"{word} is anagram with {word2}")
                anagrams.append(word2)
                if word not in anagrams:
                    anagrams.append(word)
        if not anagrams:
            anagrams.append(word)
        print("<->"*20)
        groups.append(anagrams)
    print(groups)
    
```

```python
def topKFrequent(nums, k):
    freq = {}
    for n in nums:
        freq[n] = freq.get(n, 0) + 1

    print(f"Freq -> {freq}")
    buckets = [[] for _ in range(len(nums) + 1)]
    for n, f in freq.items():
        buckets[f].append(n)
    
    print(f"Buckets -> {buckets}")

    res = []
    for f in range(len(buckets)-1, 0, -1):
        for n in buckets[f]:
            res.append(n)
            if len(res) == k:
                return res

topKFrequent(nums=nums, k=k)
```
